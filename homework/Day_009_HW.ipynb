# 檢視與處理 Outliers
# 為何會有 outliers, 常見的 outlier 原因
未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

# [作業重點]

繪製目標值累積密度函數(ECDF)的圖形, 和常態分布的累積密度函數對比, 以確認是否有離群值的情形 (In[6], Out[6], In[7], Out[7])

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
    
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data', f_app)
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
#app_train['OBS_60_CNT_SOCIAL_CIRCLE']
#dtype_select = []
#for types, col in zip(app_train.dtypes, app_train.columns):
    #print(col, types)
    #if types == 'int64' or types == 'float64':
        #dtype_select.append(col)
#print(dtype_select)

dtype_select = [np.dtype('int64'), np.dtype('float64')]
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])
print(numeric_columns)
print("Numbers of remain columns %d" % len(numeric_columns))
# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns", len(numeric_columns))
print(numeric_columns)
for col in numeric_columns:
    """
    Your CODE HERE, make the box plot
    """
    #print(app_train[col].value_counts)
    app_train[col].plot.hist(title = col)
    plt.show()
```

* REGION_POPULATION_RELATIVE有兩個值方離太遠
* DAYS_EMPLOYED有一個值方離太遠
* OWN_CAR_AGE有一個值方可能離太遠

```python
import statsmodels.api as sm
cdf = sm.distributions.ECDF(app_train['AMT_INCOME_TOTAL'])


plt.plot(cdf.x, cdf.y  / cdf.y.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([0, app_train['AMT_INCOME_TOTAL'].max()]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(cdf.x), cdf.y / cdf.y.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

app_train['AMT_INCOME_TOTAL'].plot.hist()
```

```python
# 最大值落在分布之外
print(app_train['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
Your Code Here
"""
cdf = sm.distributions.ECDF(app_train['REGION_POPULATION_RELATIVE'])


plt.plot(cdf.x, cdf.y/cdf.y.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```
